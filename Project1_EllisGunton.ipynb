```python
# Part 1: Problem Statements -- Describing the data
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv("HOF Batting Stats.csv")
df

## This dataset contains the offensive statistics of every Major League Baseball Hall of Fame player. It contains both traditional statistics (batting average, home runs, strikeouts, etc.) and more modern statistics (Wins Above Replacement, or WAR).
# Link for dataset: https://www.baseball-reference.com/awards/hof_batting.shtml
```

```python
# Part 1: Problem Statements -- Inquiry questions
df

## Main question: Are these players voted to the Hall of Fame because of a particular statistic?
# Sub-question #1: Are these players voted to the Hall of Fame because of how many home runs they hit?
# Sub-question #2: Are these players voted to the Hall of Fame because of their Wins Above Replacement (WAR)?
# Sub-question #1: Are these players voted to the Hall of Fame because of their longevity (i.e., the amount of games they played in)?
```

```python
# Part 2: Data Cleaning -- Unwanted Columns
## The only columns besides the first six I am interested in are the home run, WAR, and games played columns. I will remove the others.
cols = df.columns
remove = cols[9:15]
remove
cols[16:]
df.drop(cols[16:], inplace = True, axis = 1)
df.drop(cols[9:15], inplace = True, axis = 1)
df.drop(cols[6], inplace = True, axis = 1)
df
```

```python
# Part 2: Data Cleaning -- Missing values
print(df.isnull().sum())
# There are no missing/empty values.
```

```python
# Part 2: Data Cleaning -- Duplicate values
df = df.drop_duplicates(subset = ['Name'], keep = 'last')
df
## There are no duplicate values in this dataset.
```

```python
# Part 2: Data Cleaning -- Data types of columns
df.dtypes
# All of the data types are correct.
```

```python
# Part 2: Data Cleaning -- Inconsistent inputs
print(df['Name'].value_counts(dropna=False))
df['Name'] = df['Name'].str.lower()
df.replace()
print(df['Name'].value_counts(dropna=False))
# There appear to be no inconsistent inputs with the data.
```

```python
# Part 2: Data Cleaning -- Outliers
df.boxplot(column=['HR'])
# There appears to be no significant outliers.
```

```python
# Part 3: Question A -- Home runs
# I am looking for whether the amount of home runs hit in the players' careers affect their inclusion in the Hall of Fame. I am also looking to see the yearly trend for homers to in order to gain clues.
table1 = pd.pivot_table(df, values=['HR'], index=['Name'], columns=None, aggfunc='sum', fill_value=None, margins=False, dropna=True, margins_name='All', observed=False, sort=True)
table1 = table1.sort_values(by = 'HR', ascending = True)
print(table1)
plt.scatter(df['Inducted'], df['HR'], marker = 'o', color = 'r')
plt.show()
## According to the below graph, the number of home runs hit by a Hall of Fame baseball player increased. This follows normal trends in the history of the sport. As time went on, home runs started to become more common. The numbers of home runs started to skyrocket in the late 1980s' and early 1990s'. 
## This may have had to do with the increase in the number of players that used performance-enhancing steroids. The players in this dataset eith 100 or fewer career homers are likely pitchers (before 2022, pitchers would hit in the lineup, though not to great success) or managers.
```

```python
# Part 3: Question B -- Games played
# I am looking for whether the amount of games played in the players' careers affect their inclusion in the Hall of Fame.
table1 = pd.pivot_table(df, values=['G'], index=['Name'], columns=None, aggfunc='sum', fill_value=None, margins=False, dropna=True, margins_name='All', observed=False, sort=True)
table1 = table1.sort_values(by = 'G', ascending = True)
print(table1)
plt.scatter(df['Inducted'], df['G'], marker = 'o', color = 'r')
plt.show()
## Again, the players that played fewer than 500 games likely became Hall of Fame managers. According to the graph below, it appears the amount of games played for a Hall of Fame player slightly declines as time goes on. This may be because of the philosophy of "load management" took precedent within the last 20 years or so. 
## As a result, position players (or pitchers) would get rotated regularly in hopes of injuries becoming less prevalent. 
```

```python
# Part 3: Question C -- Wins Above Replacement
# I am looking for whether the Wins Above Replacement (WAR) in the players' careers affect their inclusion in the Hall of Fame.
table1 = pd.pivot_table(df, values=['WAR/pos'], index=['Name'], columns=None, aggfunc='sum', fill_value=None, margins=False, dropna=True, margins_name='All', observed=False, sort=True)
table1 = table1.sort_values(by = 'WAR/pos', ascending = True)
print(table1)
plt.scatter(df['Inducted'], df['WAR/pos'], marker = 'o', color = 'r')
plt.show()
## Again, the players that had an extremely low WAR (25 or lower) were likely pitchers or managers. According to the graph below, the number of players that had a WAR above 75 decreased as time went on. That is, the Hall of Fame players became more "average", so to speak, beginning around the late 1980s'. 
## However, the difference between the players with the least and most WAR in a year also get smaller. This is especially prevalent in the 2000s' when the average WAR for this period appears to be around 70 WAR compared to 80 WAR in the 1970s'.
```
